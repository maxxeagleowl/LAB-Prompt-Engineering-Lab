```python
# Prompt Engineering Lab
## Part 1: Starting Simple - Create Initial Prompts
### Step 1: Setup Your Environment
# ---

from openai import OpenAI
import os
import time
import pandas as pd

pd.set_option("display.max_colwidth", None) # Show full output in DataFrame cells

api_key = os.getenv("OPENAI_API_KEY")

# Initialize OpenAI client
client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))

def call_openai(prompt, temperature=0, max_tokens=300):
    """
    Sends a prompt to the OpenAI API and returns the model response.
    """
    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[
            {"role": "user", "content": prompt}
        ],
        temperature=temperature,
        max_tokens=max_tokens
    )
    
    return response.choices[0].message.content.strip()


def run_prompt_multiple_times(prompt, runs=5, temperature=0, max_tokens=300, delay=0.5):
    """
    Runs the same prompt multiple times and stores all responses.
    """
    results = []

    for i in range(runs):
        try:
            output = call_openai(
                prompt=prompt,
                temperature=temperature,
                max_tokens=max_tokens
            )
            results.append({
                "run": i + 1,
                "output": output
            })
            time.sleep(delay)

        except Exception as e:
            results.append({
                "run": i + 1,
                "output": f"ERROR: {e}"
            })

    return pd.DataFrame(results)


print("Setup successful. OpenAI client and helper functions are ready.")
```

```python
test_prompt = "Reply with exactly this sentence: Setup is working."

result = call_openai(test_prompt)

print(result)
```

```python
# ---
### Step 2: Create Initial Zero-Shot Prompts
# ---

# Task 1: Sentiment Analysis
sentiment_prompt_v1 = """
Classify this customer message: "I love this product! It's exactly what I needed."
"""

sentiment_result_v1 = call_openai(sentiment_prompt_v1)

print("Sentiment Analysis Result:")
print(sentiment_result_v1)


# Task 2: Product Description Generation
product_prompt_v1 = """
Create a product description for a wireless mouse that costs $29.99.
"""

product_result_v1 = call_openai(product_prompt_v1)

print("\nProduct Description Result:")
print(product_result_v1)


# Task 3: Data Extraction
extraction_prompt_v1 = """
Extract information from this customer feedback: "I ordered item #12345 on March 15th. The delivery was fast but the packaging was damaged."
"""

extraction_result_v1 = call_openai(extraction_prompt_v1)

print("\nData Extraction Result:")
print(extraction_result_v1)
```

```python
# ---
### Step 5: Run Prompts 15 Times and Create Failure Analysis
# ---

# Sentiment Analysis - 15 runs
sentiment_runs_15 = run_prompt_multiple_times(sentiment_prompt_v1, runs=15)

print("Sentiment Analysis - 15 Runs:")
print(sentiment_runs_15)


# Product Description - 15 runs
product_runs_15 = run_prompt_multiple_times(product_prompt_v1, runs=15)

print("\nProduct Description - 15 Runs:")
print(product_runs_15)


# Data Extraction - 15 runs
extraction_runs_15 = run_prompt_multiple_times(extraction_prompt_v1, runs=15)

print("\nData Extraction - 15 Runs:")
print(extraction_runs_15)
```

```python
# Create simple failure analysis summary

def analyze_consistency(df):
    unique_outputs = df["output"].nunique()
    total_runs = len(df)
    
    consistency = (1 - (unique_outputs / total_runs)) * 100
    
    return {
        "total_runs": total_runs,
        "unique_outputs": unique_outputs,
        "consistency_%": round(consistency, 2)
    }


sentiment_analysis = analyze_consistency(sentiment_runs_15)
product_analysis = analyze_consistency(product_runs_15)
extraction_analysis = analyze_consistency(extraction_runs_15)

summary_df = pd.DataFrame([
    {"task": "Sentiment Analysis", **sentiment_analysis},
    {"task": "Product Description", **product_analysis},
    {"task": "Data Extraction", **extraction_analysis}
])

print("Failure Analysis Summary:")
print(summary_df)
```
